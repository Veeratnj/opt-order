```python
def detect_bos_choch(df: pd.DataFrame):
    df = df.copy()
    df["bos"] = False
    df["choch"] = False
    df["bos_bullish"] = False
    df["bos_bearish"] = False
    df["choch_bullish"] = False
    df["choch_bearish"] = False

    # Initialize structure values
    prev_high = df["max_val"].iloc[0]
    prev_low = df["min_val"].iloc[0]
    trend = None  # "up" or "down"

    for i in range(1, len(df)):
        current_high = df.loc[i, "high"]
        current_low = df.loc[i, "low"]

        # BOS in uptrend (break previous high) → bullish BOS
        if trend in [None, "up"] and current_high > prev_high:
            df.loc[i, "bos"] = True
            df.loc[i, "bos_bullish"] = True
            trend = "up"
            prev_high = current_high

        # CHoCH in uptrend (break previous low) → bearish CHoCH
        elif trend == "up" and current_low < prev_low:
            df.loc[i, "choch"] = True
            df.loc[i, "choch_bearish"] = True
            trend = "down"
            prev_low = current_low

        # BOS in downtrend (break previous low) → bearish BOS
        elif trend == "down" and current_low < prev_low:
            df.loc[i, "bos"] = True
            df.loc[i, "bos_bearish"] = True
            prev_low = current_low

        # CHoCH in downtrend (break previous high) → bullish CHoCH
        elif trend == "down" and current_high > prev_high:
            df.loc[i, "choch"] = True
            df.loc[i, "choch_bullish"] = True
            trend = "up"
            prev_high = current_high

    return df
```

```python
def detect_bos_choch1(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    # Initialize output columns
    df["bos"] = False
    df["choch"] = False
    df["bos_bullish"] = False
    df["bos_bearish"] = False
    df["choch_bullish"] = False
    df["choch_bearish"] = False

    # Initial trend state and levels
    trend = None
    structure_high = df["high"].iloc[0]
    structure_low = df["low"].iloc[0]

    for i in range(1, len(df)):
        current_high = df.loc[i, "high"]
        current_low = df.loc[i, "low"]

        # Bullish BOS: continue or start uptrend
        if trend in [None, "up"] and current_high > structure_high:
            df.loc[i, "bos"] = True
            df.loc[i, "bos_bullish"] = True
            trend = "up"
            structure_high = current_high  # update swing high

        # Bearish CHoCH: reversal from uptrend to downtrend
        elif trend == "up" and current_low < structure_low:
            df.loc[i, "choch"] = True
            df.loc[i, "choch_bearish"] = True
            trend = "down"
            structure_low = current_low  # update swing low

        # Bearish BOS: continue downtrend
        elif trend == "down" and current_low < structure_low:
            df.loc[i, "bos"] = True
            df.loc[i, "bos_bearish"] = True
            structure_low = current_low

        # Bullish CHoCH: reversal from downtrend to uptrend
        elif trend == "down" and current_high > structure_high:
            df.loc[i, "choch"] = True
            df.loc[i, "choch_bullish"] = True
            trend = "up"
            structure_high = current_high

        # Always update structure highs/lows for the next comparison
        structure_high = max(structure_high, current_high)
        structure_low = min(structure_low, current_low)

    return df
```

```python
import mplfinance as mpf
import pandas as pd

# Set timestamp as index (needed by mplfinance)
plot_df.index = pd.to_datetime(plot_df["timestamp"])

# Prepare OHLC data
ohlc_df = plot_df[["open", "high", "low", "close", "volume"]]

# Create mplfinance addplots
apds = []

# Safely plot only if data exists (avoids zero-size array error)
if plot_df["bos_bullish"].any():
    apds.append(
        mpf.make_addplot(
            plot_df["high"].where(plot_df["bos_bullish"]).dropna(),
            type='scatter', markersize=100, marker='^', color='green'
        )
    )

if plot_df["bos_bearish"].any():
    apds.append(
        mpf.make_addplot(
            plot_df["low"].where(plot_df["bos_bearish"]).dropna(),
            type='scatter', markersize=100, marker='v', color='darkred'
        )
    )

if plot_df["choch_bullish"].any():
    apds.append(
        mpf.make_addplot(
            plot_df["high"].where(plot_df["choch_bullish"]).dropna(),
            type='scatter', markersize=100, marker='^', color='cyan'
        )
    )

if plot_df["choch_bearish"].any():
    apds.append(
        mpf.make_addplot(
            plot_df["low"].where(plot_df["choch_bearish"]).dropna(),
            type='scatter', markersize=100, marker='v', color='orange'
        )
    )

# Add max_val and min_val lines only if valid
if plot_df["max_val"].notna().any():
    apds.append(mpf.make_addplot(plot_df["max_val"].dropna(), color='red', linestyle='dotted'))

if plot_df["min_val"].notna().any():
    apds.append(mpf.make_addplot(plot_df["min_val"].dropna(), color='blue', linestyle='dotted'))

# Final candlestick plot
mpf.plot(
    ohlc_df,
    type='candle',
    style='yahoo',
    title='BOS & CHoCH - Top 100 Candles',
    ylabel='Price',
    addplot=apds,
    volume=False,
    figratio=(16, 8),
    figscale=1.2,
    tight_layout=True
)
```
